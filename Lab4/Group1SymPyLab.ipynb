SymPy’s polynomials

```python
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**3 - 8*x**2  + x - 18

p, r = div(p,  x-1)

print(p)
print(r)

p, r = div(p,  x-2)

print(p)
print(r)

p, r = div(p,  x-3)

print(p)
print(r)
```

```python
p = 2*x**2 + 6*x - 12

p, r = div(p,  x-3)

print(p)
print(r)
```

SymPy’s polynomial simple univariate polynomial factorization

```python
from sympy import Symbol,factor
x = Symbol('x')
factor(7*x**4 - 35*x**3 + 21*x)
```

SymPy’s solvers

```python
from sympy import Symbol, solveset, Eq
x = Symbol('x')
solveset(Eq(x**4 - 3*x**3  - 21*x**2 + 81*x + 270, 0), x)
```

```python
from sympy import Symbol, solveset, Eq, pi
x = Symbol('x')
solveset(Eq(x**pi + ((2)-pi)*x + (pi) , 0), x)
```

```python
from sympy import Symbol,factor
x = Symbol('y')
factor( x**4  + 4*x**3 + 8*x**2 - 16*x,domain='GF(2)')
```

SymPy’s Symbolic and Numerical Complex Evaluations

```python
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = z2 + I*x2
y = y2 + I*z1
z = z1 + I*y1

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))

w=N(5/(I), 20)
print('w=',w)
```

SymPy’s integrate

```python
from sympy import Symbol, integrate, sin
x = Symbol('x')
integrate(x**3 +x**5, x)
```

```python
from sympy import Symbol, integrate, sin, cos, pi
x = Symbol('x')
integrate(sin(x) * cos(x**2), (x,pi,2*pi))
```

```python
from sympy import Symbol, integrate, N
x = Symbol('x')
N(integrate(x**2 - x**3**x, (x,1,5)))
```
